```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
```

```python
sensor=pd.read_csv('sensor.csv' ,header=0 ,parse_dates=[1])
```

```python
sensor.head(3)
```

```python
sensor.drop('Unnamed: 0',axis=1 ,inplace=True)
```

```python
sensor.fillna(0)
```

```python
sensor['machine_status'].unique()
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
sensor['machine_status']=label_encoder.fit_transform(sensor['machine_status'])
```

### Time series 

```python
timeseries=sensor[['timestamp','machine_status']]
```

```python
timeseries['Monthly']=timeseries['timestamp'].dt.month
```

```python
timeseries['machine_status'].plot.bar()
```
